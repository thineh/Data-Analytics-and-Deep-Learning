### CÂU 13:

```python
mean_age = df['Age'].mean()
df.loc[df['Age'] > 40, 'Age'] = mean_age
print(df)
```

### CÂU 14:

```python
df['cs_BMI'] = df['Weight'] / (df['Height'] ** 2)
print(df)
```

### CÂU 15:

```python
def phan_loai_bmi(bmi):
    if bmi < 18.5:
        return 0  # Thiếu cân
    elif 18.5 <= bmi < 23:
        return 1  # Bình thường
    elif 23 <= bmi < 25:
        return 2  # Thừa cân
    else:
        return 3  # Béo phì

df['Phan_Loai'] = df['cs_BMI'].apply(phan_loai_bmi)
print(df)
```

### CÂU 16:

```python
df['cs_BMI_normalized'] = (df['cs_BMI'] - df['cs_BMI'].min()) / (df['cs_BMI'].max() - df['cs_BMI'].min())
print(df)
```

### CÂU 17:

```python
def get_GT(name):
    if 'Mr.' in name:
        return 'male'
    elif 'Mrs.' in name or 'Miss.' in name:
        return 'female'
    else:
        return None

df['GT'] = df['Name'].apply(get_GT)
df['Name'] = df['Name'].str.replace('Mr.|Mrs.|Miss.', '', regex=True).str.strip()
print(df)
```

### CÂU 18:

```python
GT_counts = df['GT'].value_counts()
plt.pie(GT_counts, labels=GT_counts.index, autopct='%1.1f%%')
plt.title('Biểu đồ tròn cho GT')
plt.show()
```

### CÂU 19:

```python
female_data = df[df['GT'] == 'female']['Phan_Loai'].value_counts()
plt.bar(female_data.index, female_data.values)
plt.xticks([0, 1, 2, 3], ['Thiếu cân', 'Bình thường', 'Thừa cân', 'Béo phì'])
plt.xlabel('Phân loại BMI')
plt.ylabel('Số lượng')
plt.title('Phân loại BMI của người nữ')
plt.show()
```
